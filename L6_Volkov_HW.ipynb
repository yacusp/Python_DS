Импортируйте библиотеки pandas и numpy.

Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. 
Создайте датафреймы X и y из этих данных.

Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки
составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.


```python
# импорт библиотек
import pandas as pd
import numpy as np

# загрузка набора данных
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
data = boston["data"]
```

```python
feature_names = boston["feature_names"]

# создание X датафрейма
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
target = boston["target"]

# создание Y датафрейма
Y = pd.DataFrame(target, columns=["price"])
Y.head()
```

```python
# разбиваем датафреймы на тренировочные и тестовые
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
# построение модели линейной регрессии
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# обучение модели
lr.fit(X_train, Y_train)
```

```python
# получение предсказанных значений
Y_pred = lr.predict(X_test)
```

```python
# сравнение полученных значений с тестовыми
check_test = pd.DataFrame({
    "Y_test": Y_test["price"],
    "Y_pred": Y_pred.flatten(),
})

check_test.head(10)
```

```python
# вычисление R2 полученных предсказаний
from sklearn.metrics import r2_score

r2_score(Y_test, Y_pred)
```

```python
# подключение библиотек 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
# загрузка файла
df = pd.read_csv('creditcard.zip', compression='zip')
```

```python
# проверка датафрейма 
df.value_counts(normalize=True)
```

```python
# проверка стобцов методом info
df.info()
```

```python
# просмотр всех столбцов
pd.options.display.max_columns=100

df.head(10)
```

```python
# создание датафреймов X и y
X = df.drop('Class', axis=1)

y = df['Class']
```

```python
# разбивка на тестовые и тренеровочные наборы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
# проверка формы
print('X_train ', X_train.shape)
print('X_test ', X_test.shape)
print('y_train ', y_train.shape)
print('y_test ', y_test.shape)
```

```python
# залаём параметры
parameters = [{
    'n_estimators': [10, 15], 
    'max_features': np.arange(3, 5), 
    'max_depth': np.arange(4, 7)
}] 
```

```python
# создание модели
gsc = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc', cv=3
    )
```

```python
# обучение модели
gsc.fit(X_train, y_train)
```

```python
# просмотр параметров
gsc.best_params_
```

```python
# предсказание вероятности
y_pred = gsc.predict_proba(X_test)
```

```python
# выбор столбца с индексом 1 и вычисление AUC
y_pred_proba = y_pred[:, 1]
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
